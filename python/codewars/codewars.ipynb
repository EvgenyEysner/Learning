Ava, Mark, Sheila und Pete sind auf einer Party. Ava und Sheila bleiben jedoch nur, wenn mindestens 4 Personen anwesend sind, Pete bleibt nur, wenn mindestens 1 Person anwesend ist, und Mark bleibt nur, wenn mindestens 5 Personen anwesend sind. Daher geht Mark, woraufhin auch Ava und Sheila gehen, und Pete bleibt allein zurück.
Gegeben sei ein Array mit den Präferenzen jeder Person auf einer Party hinsichtlich der Mindestanzahl an anwesenden Personen. Bestimme, wie viele Personen bleiben werden.
Beispiele

```
[4, 5, 4, 1] ➞ 1
# Avas Mindestanzahl ist 4, Marks ist 5, Sheilas ist 4 und Petes ist 1.
# Nur 1 Person (Pete) bleibt.

[10, 12, 15, 15, 5] ➞ 0

[2, 1, 2, 0] ➞ 4

```
Hinweise

* Alle Teilnehmer sind im Array enthalten.
* Die Anzahl der Personen, die sich derzeit auf der Party befinden, schließt die jeweilige Person selbst mit ein.
* Die Mindestanzahl einer Person kann 0, 1 oder eine beliebige positive ganze Zahl sein.
* Es wird nur gültige Eingabe erwartet.
* 0 <= len(lst) <= 10^5


````python
def party(lst):
    lst.sort()
    for k in range(len(lst), 0, -1):
        if lst[k - 1] <= k:
            return k
    return 0

party([4, 5, 4, 1])
````

Kernidee: Sortiere die Präferenzen aufsteigend. Dann prüfe von groß nach klein: Bleiben k Personen, müssen die k mit den niedrigsten Anforderungen alle ≤ k haben.
Mit sortierten Werten `[1, 4, 4, 5] und k=4 Personen:

Die 4 Personen mit den niedrigsten Anforderungen bleiben am ehesten
lst[k-1]` ist die höchste Anforderung unter diesen **k** Personen
Wenn diese **≤ k** → alle k sind zufrieden → stabiler Zustand ✓

```Python
[4,5,4,1] → sortiert: [1,4,4,5]
  k=4: lst[3]=5 > 4 ✗
  k=3: lst[2]=4 > 3 ✗
  k=2: lst[1]=4 > 2 ✗
  k=1: lst[0]=1 ≤ 1 ✓ → return 1
##############################################
Tests:
pythonprint(party([4, 5, 4, 1]))   # → 1
print(party([10,12,15,15,5]))# → 0
print(party([2, 1, 2, 0]))   # → 4
print(party([]))              # → 0
```
Zeitkomplexität: O(n log n) durch Sortierung — effizient auch für 10⁵ Elemente.
