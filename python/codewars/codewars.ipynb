Ava, Mark, Sheila und Pete sind auf einer Party. Ava und Sheila bleiben jedoch nur, wenn mindestens 4 Personen anwesend sind, Pete bleibt nur, wenn mindestens 1 Person anwesend ist, und Mark bleibt nur, wenn mindestens 5 Personen anwesend sind. Daher geht Mark, woraufhin auch Ava und Sheila gehen, und Pete bleibt allein zurück.
Gegeben sei ein Array mit den Präferenzen jeder Person auf einer Party hinsichtlich der Mindestanzahl an anwesenden Personen. Bestimme, wie viele Personen bleiben werden.
Beispiele

```
[4, 5, 4, 1] ➞ 1
# Avas Mindestanzahl ist 4, Marks ist 5, Sheilas ist 4 und Petes ist 1.
# Nur 1 Person (Pete) bleibt.

[10, 12, 15, 15, 5] ➞ 0

[2, 1, 2, 0] ➞ 4

```
Hinweise

* Alle Teilnehmer sind im Array enthalten.
* Die Anzahl der Personen, die sich derzeit auf der Party befinden, schließt die jeweilige Person selbst mit ein.
* Die Mindestanzahl einer Person kann 0, 1 oder eine beliebige positive ganze Zahl sein.
* Es wird nur gültige Eingabe erwartet.
* 0 <= len(lst) <= 10^5


````python
def party_people(lst):
    lst.sort()
    for k in range(len(lst), 0, -1):
        if lst[k - 1] <= k:
            return k
    return 0

party_people([4, 5, 4, 1])
````

Kernidee: Sortiere die Präferenzen aufsteigend. Dann prüfe von groß nach klein: Bleiben k Personen, müssen die k mit den niedrigsten Anforderungen alle ≤ k haben.
Mit sortierten Werten `[1, 4, 4, 5] und k=4 Personen:

Die 4 Personen mit den niedrigsten Anforderungen bleiben am ehesten
lst[k-1]` ist die höchste Anforderung unter diesen **k** Personen
Wenn diese **≤ k** → alle k sind zufrieden → stabiler Zustand ✓

```Python
[4,5,4,1] → sortiert: [1,4,4,5]
  k=4: lst[3]=5 > 4 ✗
  k=3: lst[2]=4 > 3 ✗
  k=2: lst[1]=4 > 2 ✗
  k=1: lst[0]=1 ≤ 1 ✓ → return 1
##############################################
Tests:
pythonprint(party([4, 5, 4, 1]))   # → 1

print(party([2, 1, 2, 0]))   # → 4
print(party([]))              # → 0
```
Zeitkomplexität: O(n log n) durch Sortierung — effizient auch für 10⁵ Elemente.

Aufgabe
Manche Zahlen haben interessante Eigenschaften. Zum Beispiel:

89 → 8¹ + 9² = 89 (also 89 × 1)
695 → 6² + 9³ + 5⁴ = 1390 (also 695 × 2)
46288 → 4³ + 6⁴ + 2⁵ + 8⁶ + 8⁷ = 2360688 (also 46288 × 51)


Du bekommst zwei positive Zahlen: eine Zahl n und einen Startexponenten p.
Nimm jede Ziffer von n und erhöhe sie auf aufsteigende Potenzen, beginnend bei p. Addiere alle Ergebnisse.
Prüfe dann: Ist diese Summe ein ganzzahliges Vielfaches von n?

Ja → gib den Faktor k zurück
Nein → gib -1 zurück
Beispiele:
```
n = 89, p = 1
8¹ + 9² = 8 + 81 = 89
89 / 89 = 1  →  Rückgabe: 1

n = 92, p = 1
9¹ + 2² = 9 + 4 = 13
13 / 92 → kein ganzzahliges Ergebnis  →  Rückgabe: -1

n = 695, p = 2
6² + 9³ + 5⁴ = 36 + 729 + 625 = 1390
1390 / 695 = 2  →  Rückgabe: 2
```

````python
def dig_pow(n, p):
    digits = [int(d) for d in str(n)]
    total = sum(d**(p+i) for i, d in enumerate(digits))
    return total // n if total % n == 0 else -1
````

ERklärung:
```python

# 695 → ['6','9','5'] → [6, 9, 5]

# i=0: 6^(2+0) = 6² = 36
# i=1: 9^(2+1) = 9³ = 729
# i=2: 5^(2+2) = 5⁴ = 625
# total = 1390

return total // n if total % n == 0 else -1
# 1390 % 695 == 0 → 1390 // 695 = 2 ✓
```
`enumerate` ist der Schlüssel:
```python
enumerate([6, 9, 5])
# → (0, 6), (1, 9), (2, 5)
#     ↑ i      ↑ i      ↑ i
# Exponent = p + i → steigt automatisch mit jeder Ziffer
```
Kurz gesagt: `enumerate` liefert den Index i, der auf den Startexponenten p addiert wird — so bekommt jede Ziffer automatisch den nächsten Exponenten.

Schreibe eine Funktion namens `setAlarm/set_alarm/set-alarm/setalarm (je nach Programmiersprache)`, die zwei Parameter entgegennimmt. Der erste Parameter, `employed`, ist wahr, wenn du berufstätig bist, und der zweite Parameter, `vacation`, ist wahr, wenn du im Urlaub bist.

Die Funktion sollte „true“ zurückgeben, wenn du berufstätig bist und nicht im Urlaub (da dies die Umstände sind, unter denen du einen Wecker stellen musst). Andernfalls sollte sie `false` zurückgeben.
### Beispiele:
```python
employed | vacation
true     | true     => false
true     | false    => true
false    | true     => false
false    | false    => false
```

````python
def set_alarm(employed, vacation):
    return employed and not vacation
````

Wenn wir alle natürlichen Zahlen unter 10 auflisten, die Vielfache (teilbar durch 3 oder 5) von 3 oder 5 sind, erhalten wir 3, 5, 6 und 9. Die Summe dieser Vielfachen beträgt 23.
Vervollständigen Sie die Lösung so, dass sie die Summe aller Vielfachen von 3 oder 5 unterhalb der übergebenen Zahl zurückgibt.
Wenn die Zahl negativ ist, geben Sie außerdem 0 zurück.

Hinweis: Wenn eine Zahl sowohl ein Vielfaches von 3 als auch von 5 ist, zählen Sie sie nur einmal.


````python
def solution(number):
    if number < 0:
        return 0
    return sum(i for i in range(number) if i % 3 == 0 or i % 5 == 0)
````

Pete backt gerne Kuchen. Er hat einige Rezepte und Zutaten. Leider ist er nicht gut in Mathe. Kannst du ihm helfen herauszufinden, wie viele Kuchen er mit seinen Rezepten backen könnte?
Schreibe eine Funktion `cakes()`, die das Rezept (Objekt) und die verfügbaren Zutaten (ebenfalls ein Objekt) als Argumente entgegennimmt und die maximale Anzahl an Kuchen zurückgibt, die Pete backen kann (Ganzzahl). Der Einfachheit halber gibt es keine Einheiten für die Mengen (z. B. sind 1 lb Mehl oder 200 g Zucker einfach 1 oder 200). Zutaten, die in den Objekten nicht vorhanden sind, können als 0 betrachtet werden.

Beispiele:
```python
# must return 2
cakes({flour: 500, sugar: 200, eggs: 1}, {flour: 1200, sugar: 1200, eggs: 5, milk: 200})
# must return 0
cakes({apples: 3, flour: 300, sugar: 150, milk: 100, oil: 100}, {sugar: 500, flour: 2000, milk: 2000})
```

````python
def cakes(recipe, available):
    return min(available.get(k, 0) // recipe[k] for k in recipe)

cakes({"flour": 500, "sugar": 200, "eggs": 1}, {"flour": 1200, "sugar": 1200, "eggs": 5, "milk": 200})

cakes({"apples": 3, "flour": 300, "sugar": 150, "milk": 100, "oil": 100}, {"sugar": 500, "flour": 2000, "milk": 2000})
````

Schreibe eine Funktion, die eine Zeichenkette als Eingabe entgegennimmt und das erste Zeichen zurückgibt, das an keiner Stelle in der Zeichenkette wiederholt vorkommt.
Wenn beispielsweise die Eingabe `stress` gegeben ist, sollte die Funktion `t` zurückgeben, da der Buchstabe `t` nur einmal in der Zeichenkette vorkommt und an erster Stelle steht.
Als zusätzliche Herausforderung gelten Groß- und Kleinbuchstaben als dasselbe Zeichen, die Funktion sollte jedoch die richtige Schreibweise für das Anfangszeichen zurückgeben. Bei der Eingabe „sTreSS“ sollte beispielsweise „T“ zurückgegeben werden.

Wenn eine Zeichenkette nur sich wiederholende Zeichen enthält, gib eine leere Zeichenkette ("") zurück.
Hinweis: Ungeachtet des Namens in einigen Sprachen sollte deine Funktion alle Unicode-Codepunkte verarbeiten:
```python
"@#@@*"    --> "#"
"かか何"   --> "何"
"🐐🦊🐐" --> "🦊"
```

````python
def first_non_repeating_letter(string):
    lower_string = string.lower()
    for i, char in enumerate(lower_string):
        if lower_string.count(char) == 1:
            return  string[i]
    return ""
first_non_repeating_letter('a')
first_non_repeating_letter('stress')
first_non_repeating_letter('moonmen')
````
