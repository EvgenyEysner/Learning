1. Prüfen Sie, ob die angegebene Zahl gerade ist oder nicht. Ihre Funktion sollte True zurückgeben, wenn die Zahl gerade ist, und False, wenn die Zahl ungerade ist. <br/>
    Eingabe: Eine ganze Zahl.

    Ausgabe: Bool.

    Beispiel:
    <hr/>
    assert is_even(2) == True <br/>
    assert is_even(5) == False <br/>
    assert is_even(0) == True <br/>

````python
def is_even(num: int) -> bool:
    return num & 1 == 0


# These "asserts" are used for self-checking and not for an auto-testing
assert is_even(2) == True
assert is_even(5) == False
assert is_even(0) == True
````

Diese Funktion `is_even` prüft, ob eine gegebene ganze Zahl gerade ist. Hier ist die Erklärung Schritt für Schritt:

### **1. Funktionssignatur**
```python
def is_even(num: int) -> bool:
```
- **`def is_even(...):`** Definiert eine Funktion namens `is_even`.
- **`num: int`** Der Parameter `num` ist eine ganze Zahl (`int`).
- **`-> bool`** Die Funktion gibt einen Boolean-Wert (`True` oder `False`) zurück.

---

### **2. Bitweiser AND-Operator (`&`)**
```python
return num & 1 == 0
```
- **`num & 1`**:
  - Der bitweise AND-Operator vergleicht die Binärzahlen von `num` und `1`.
  - In Binärform ist `1` immer `0b1` (also nur das letzte Bit gesetzt).
  - Beispiel:
    - Für `num = 4` (Binär: `0b100`) → `4 & 1 = 0`
    - Für `num = 5` (Binär: `0b101`) → `5 & 1 = 1`

- **Ergebnis**:
  - Wenn das **letzte Bit** von `num` `0` ist (gerade Zahl), ergibt `num & 1` den Wert `0`.
  - Wenn das letzte Bit `1` ist (ungerade Zahl), ergibt `num & 1` den Wert `1`.

---

### **3. Vergleich mit `== 0`**
- **`num & 1 == 0`** prüft, ob das Ergebnis des bitweisen ANDs gleich `0` ist.
  - Ist dies der Fall, ist die Zahl **gerade** → Rückgabewert `True`.
  - Andernfalls **ungerade** → Rückgabewert `False`.

---

### **4. Warum funktioniert das?**
- Gerade Zahlen haben in ihrer Binärdarstellung immer eine `0` als letztes Bit.
- Ungerade Zahlen haben immer eine `1` als letztes Bit.
- Durch das AND mit `1` wird nur das letzte Bit überprüft.

---

### **5. Vergleich mit `num % 2 == 0`**
- Alternativ könnte man auch `num % 2 == 0` verwenden, um Geradheit zu prüfen.
- Beide Methoden sind korrekt, aber:
  - `num & 1 == 0` ist **etwas schneller**, da bitweise Operationen in der Regel effizienter sind.
  - `num % 2 == 0` ist **lesbarer**, besonders für Einsteiger.

---

### **6. Beispiele**
```python
print(is_even(4))  # True  (4 & 1 = 0)
print(is_even(5))  # False (5 & 1 = 1)
print(is_even(0))  # True  (0 & 1 = 0)
print(is_even(-2)) # True  (-2 & 1 = 0)
```

---

### **Zusammenfassung**
Die Funktion nutzt bitweise Logik, um schnell und effizient zu prüfen, ob eine Zahl gerade ist. Sie ist besonders in Performance-kritischen Anwendungen nützlich, aber für den Alltag ist `num % 2 == 0` oft besser verständlich.

2. Sie haben eine Folge von Zeichenfolgen und möchten die am häufigsten vorkommende Zeichenfolge in der Folge ermitteln. Es kann die einzige sein.

    Beispiel:
    <hr/>
    Eingabe: Nicht-leeres Liste (list) von String (str).<br/>
    Ausgabe: Eine Zeichenkette (str).<br/>

    Beispiele:
    <hr/>
    assert most_frequent(["a", "b", "c", "a", "b", "a"]) == "a" <br/>
    assert most_frequent(["a", "a", "bi", "bi", "bi"]) == "bi"

````python
from collections import Counter


def most_frequent(data: list[str]) -> str:
    return Counter(data).most_common(1)[0][0]

# These "asserts" are used for self-checking
assert most_frequent(["a", "b", "c", "a", "b", "a"]) == "a"
assert most_frequent(["a", "a", "bi", "bi", "bi"]) == "bi"
````

1. Counter(strings): Erstellt ein Zählerobjekt, das die Häufigkeit jeder Zeichenfolge in der Eingabefolge speichert.

2. most_common(1): Gibt eine Liste der häufigsten Elemente zurück. Der Parameter 1 begrenzt die Ausgabe auf das häufigste Element.

3. [0][0]: Greift auf das erste Element des zurückgegebenen Tupels (Zeichenfolge, Häufigkeit) zu.

12. Aufgabe: In einer gegebenen Sequenz sollte das erste Element das letzte sein. Eine leere Sequenz oder eine Sequenz mit nur einem Element sollte unverändert bleiben.
    Eingabe: Liste (List).
    Ausgabe: Liste (List) oder ein anderes Iterable (Tupel (tuple), Iterator (iterator), Generator (generator)).

    assert replace_first([1, 2, 3, 4]) == [2, 3, 4, 1]
    assert replace_first([1]) == [1]
    assert replace_first([]) == []

````python
from collections.abc import Iterable


def replace_first(items: list) -> Iterable:
    if len(items) <= 1:
        return items
    return items[1:] + [items[0]]

replace_first([1, 2, 3, 4, 5])
````


13. Sie haben eine Zeichenkette, die nur aus Ziffern besteht. Sie müssen herausfinden, wie viele Nullen ("0") am Anfang der gegebenen Zeichenkette stehen.

Eingabe: Eine Zeichenkette (str), die aus Ziffern besteht.

Ausgabe: Eine ganze Zahl (int).

Beispiele:

1. assert beginning_zeros("100") == 0
2. assert beginning_zeros("001") == 2
3. assert beginning_zeros("100100") == 0
4. assert beginning_zeros("001001") == 2

Vorbedingung: 0-9

````python
def beginning_zeros(a: str) -> int:
    count = 0
    for char in a:
        if char == '0':
            count += 1
        else:
            break
    
    return int(count)

assert beginning_zeros("100") == 0
assert beginning_zeros("001") == 2
assert beginning_zeros("100100") == 0
assert beginning_zeros("001001") == 2
````

14. Sie erhalten eine Zeichenkette und zwei Markierungen (die erste und die letzte). Sie müssen eine Teilzeichenkette finden, die zwischen diesen beiden Markierungen liegt. Aber es gibt ein paar wichtige Bedingungen.

Die Anfangs- und die Endmarkierung sind immer unterschiedlich.
Die Anfangs- und die Endmarkierung sind immer 1 Zeichen lang.
Die Anfangs- und Endmarkierungen befinden sich immer in einer Zeichenkette und folgen aufeinander.
example

Eingabe: Drei Argumente. Alle sind Zeichenketten (str) . Das zweite und dritte Argument sind die Anfangs- und Endmarkierungen.

Ausgabe: Eine Zeichenkette (str) .

Beispiele:

1. assert between_markers("What is >apple<", ">", "<") == "apple"
2. assert between_markers("What is [apple]", "[", "]") == "apple"
3. assert between_markers("What is ><", ">", "<") == ""
4. assert between_markers("[an apple]", "[", "]") == "an apple"

Wie es verwendet wird: Zum Parsen von Text.
Vorbedingung: Es darf nicht mehr als eine End- und eine Anfangsmarkierung geben.



````python
def between_markers(text: str, start: str, end: str) -> str:
    start_index = text.find(start)
    end_index = text.find(end, start_index + len(start))
    if start_index != -1 and end_index != -1:
        return text[start_index + len(start):end_index]
    return ""


assert between_markers("What is >apple<", ">", "<") == "apple"
assert between_markers("What is [apple]", "[", "]") == "apple"
assert between_markers("What is ><", ">", "<") == ""
assert between_markers("[an apple]", "[", "]") == "an apple"
````

Teilen Sie die Zeichenfolge in Paare von zwei Zeichen auf. Wenn die Zeichenfolge eine ungerade Anzahl von Zeichen enthält, sollte das fehlende zweite Zeichen des letzten Paares durch einen Unterstrich ('_') ersetzt werden.

Eingabe: Eine Zeichenkette.

Ausgabe: Eine Liste oder ein anderes Iterable von Strings.

Beispiel:

1. assert list(split_pairs("abcd")) == ["ab", "cd"]
2. assert list(split_pairs("abc")) == ["ab", "c_"]
3. assert list(split_pairs("abcdf")) == ["ab", "cd", "f_"]
4. assert list(split_pairs("a")) == ["a_"]

Vorbedingung: 0 <= len(text) <= 100

````python
from itertools import zip_longest
from typing import List
from typing import Iterable


# 1. Möglichkeit 
def split_pairs(text: str) -> List[str]:
    pairs = zip_longest(text[::2], text[1::2], fillvalue='_')
    return [a + b for a, b in pairs]

# 2. Möglichkeit
def split_pairs(text: str) -> Iterable[str]:
    for i in range(0, len(text), 2):
        print(text[i:i + 2])
        yield text[i:i + 2].ljust(2, '_')

assert list(split_pairs("abcd")) == ["ab", "cd"]
assert list(split_pairs("abc")) == ["ab", "c_"]
assert list(split_pairs("abcdf")) == ["ab", "cd", "f_"]
assert list(split_pairs("a")) == ["a_"]
````

Als Eingabe für Ihre Funktion wird Ihnen ein Satz vorgegeben. Sie müssen eine korrigierte Version zurückgeben, die mit einem Großbuchstaben beginnt und mit einem Punkt endet.

Achten Sie darauf, dass nicht alle Korrekturen notwendig sind. Wenn ein Satz bereits mit einem Punkt endet, ist das Hinzufügen eines weiteren Punktes ein Fehler.

Eingabe: Eine Zeichenkette (str).

Ausgabe: Eine Zeichenkette (str).

Beispiele:

1. assert correct_sentence("greetings, friends") == "Greetings, friends."
2. assert correct_sentence("Greetings, friends") == "Greetings, friends."
3. assert correct_sentence("Greetings, friends.") == "Greetings, friends."
4. assert correct_sentence("greetings, friends.") == "Greetings, friends."

Vorbedingung: Keine führenden und nachgestellten Leerzeichen, Text enthält nur Leerzeichen, a-z, A-Z, "," und ".".

````python
def correct_sentence(text: str) -> str:
    if text and not text[0].isupper():
        text = text[0].upper() + text[1:]

    if text and not text.endswith('.'):
        text += '.'
    
    return text


print("Example:")
print(correct_sentence("greetings, friends"))

# These "asserts" are used for self-checking
assert correct_sentence("greetings, friends") == "Greetings, friends."
assert correct_sentence("Greetings, friends") == "Greetings, friends."
assert correct_sentence("Greetings, friends.") == "Greetings, friends."
assert correct_sentence("greetings, friends.") == "Greetings, friends."

print("The mission is done! Click 'Check Solution' to earn rewards!")
````

Finden Sie den Wert, der dem angegebenen Wert am nächsten kommt.

Sie erhalten eine Menge von ganzen Zahlen und einen Wert, für den Sie den nächstliegenden Wert finden müssen.

Wir haben zum Beispiel die folgende Zahlenfolge: 4, 7, 10, 11, 12, 17, und wir müssen den Wert finden, der der Zahl 9 am nächsten liegt. Wenn wir diese Folge sortieren aufsteigend sortieren, dann wird links von der Zahl 9 die Zahl 7 und rechts davon die Zahl 10 stehen rechts - die Zahl 10. Aber 10 liegt näher als 7, was bedeutet, dass die richtige Antwort 10 ist.

Ein paar Klarstellungen:

Wenn 2 Zahlen den gleichen Abstand haben, müssen Sie die kleinste Zahl wählen;
Die Zahlenfolge ist immer nicht leer;
Der angegebene Wert kann in dieser Folge enthalten sein, was bedeutet, dass er die Antwort ist;
Die Folge kann sowohl positive als auch negative Zahlen enthalten, aber es handelt sich immer um ganze Zahlen;
Die Folge ist nicht sortiert und besteht nur aus eindeutigen Zahlen.
Eingabe: Zwei Argumente. Eine Menge von ganzen Zahlen. Der gesuchte Wert als eine ganze Zahl.

Ausgabe: eine ganze Zahl.

Beispiele:

​assert nearest_value({17, 4, 7, 10, 11, 12}, 9) == 10
assert nearest_value({17, 4, 7, 10, 11, 12}, 8) == 7
assert nearest_value({17, 4, 8, 10, 11, 12}, 9) == 8
assert nearest_value({17, 4, 9, 10, 11, 12}, 9) == 9

````python
def nearest_value(values: set, target: int) -> int:
    # Einzeiler mit min() Funktion und custom key
    return min(values, key=lambda x: (abs(x - target), x))


print("Example:")
print(nearest_value({4, 7, 10, 11, 12, 17}, 9))

assert nearest_value({17, 4, 7, 10, 11, 12}, 9) == 10
assert nearest_value({17, 4, 7, 10, 11, 12}, 8) == 7
assert nearest_value({17, 4, 8, 10, 11, 12}, 9) == 8
assert nearest_value({17, 4, 9, 10, 11, 12}, 9) == 9
assert nearest_value({17, 4, 7, 10, 11, 12}, 0) == 4
assert nearest_value({17, 4, 7, 10, 11, 12}, 100) == 17
assert nearest_value({100, 5, 8, 89, 10, 12}, 7) == 8
assert nearest_value({2, 3, -1}, 0) == -1
assert nearest_value({5}, 5) == 5
assert nearest_value({5}, 7) == 5

print("The mission is done! Click 'Check Solution' to earn rewards!")
````
